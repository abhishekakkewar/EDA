```python
df_country=pd.read_excel("Country-Code.xlsx")
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
county_name = final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val,labels=county_name)
```

```python
#Top 3 countris that uses zomato
plt.pie(country_val[:3],labels=county_name[:3], autopct='%1.2f%%')
```

```python
# Observation: Zomato maximum transaction are from India After that USA and then UK
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
##Conclusion
```

#Observation
1. when rating is between 4.5 t0 4.9 -> excellent
2. when rating are betwween 4.0 to 4.4 -> very good
3. 3.5 to 3.9 -> good
4. 3.0 - 3.4 -> Avg
5. 2.5 - 2.9  -> Avg
6. 2.0 - 2.4 -> Poor
